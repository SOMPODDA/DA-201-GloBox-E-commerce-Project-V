```python
## 95% CONFIDNECE INTERVAL FOR THE DIFFERENCE IN THE AMOUNT SPENT PER USER BETWEEN THE TREATMENT AND THE CONTROL(treatment-control) 
## USING THE t DISTRIBUTION AND ASSUMING UNEQUAL VARIANCE ##
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Define the sample statistics and parameters
control_mean = 3.375
control_std = 25.94
control_n = 24343
treatment_mean = 3.391
treatment_std = 25.41
treatment_n = 24600
df = control_n + treatment_n - 2
alpha = 0.05
se = np.sqrt((control_std**2)/control_n + (treatment_std**2)/treatment_n)
t_crit = t.ppf(1 - alpha/2, df)
me = t_crit * se
diff = treatment_mean - control_mean
lower = diff - me
upper = diff + me

# Set colorblind-friendly colors
colors = ['#1f77b4', '#ff7f0e']

# Plot the probability distribution
x = np.linspace(-1, 1, 500)
y = t.pdf(x, df, diff, se)
plt.plot(x, y, label='Probability Distribution', color=colors[0])

# Shade the area corresponding to the confidence interval
x_ci = np.linspace(lower, upper, 500)
y_ci = t.pdf(x_ci, df, diff, se)
plt.fill_between(x_ci, y_ci, 0, alpha=0.2, label='95% Confidence Interval', color=colors[1])

# Add labels and legend
plt.xlabel('Difference in Means')
plt.ylabel('Probability Density')
plt.title('Probability Distribution with 95% Confidence Interval')

# Add the bounds to the plot
plt.axvline(lower, color='red', linestyle='--', label=f'Lower Bound = {lower:.10f}')
plt.axvline(upper, color='red', linestyle='--', label=f'Upper Bound = {upper:.10f}')

# Additional insights
plt.text(0, 0.15, 'Positive Difference in Means', color='black', ha='center')

# Display the legend
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

```python
## HYPOTHESIS TEST TO SEE WHETHER THERE IS A DIFFERENCE IN THE CONVERSION RATE BETWEEN THE TWO GROUPS. 
## USING THE NORMAL DISTRIBUTION AT A 5% SIGNIFICANCE LEVEL AND THE POOLED PROPORTION FOR THE STANDARD ERROR ##
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the p-value
p_value = 0.000111556
# Define the significance level
significance_level = 0.05
# Define the test statistic
test_statistic = -3.864291770

# Generate x values for the standard normal distribution
x = np.linspace(-5, 5, 100)

# Calculate the corresponding y values from the standard normal distribution
y = norm.pdf(x, 0, 1)

# Plot the standard normal distribution
plt.plot(x, y, color='navy', label='Standard Normal Distribution')

# Shade the area representing the p-value
plt.fill_between(x, y, where=(x <= test_statistic), color='skyblue', alpha=0.4, label='p-value')

# Shade the area representing the acceptance region (null hypothesis accepted)
plt.fill_between(x, y, where=(x >= -norm.ppf(1 - significance_level)), color='lightgreen', alpha=0.4, label='Acceptance Region')

# Shade the area representing the rejection region (null hypothesis rejected)
plt.fill_between(x, y, where=((x < -norm.ppf(significance_level/2)) & (x > -norm.ppf(1 - significance_level/2))), color='salmon', alpha=0.4, label='Rejection Region')

# Add vertical lines for the significance level and test statistic
plt.axvline(-norm.ppf(significance_level/2), color='red', linestyle='--', label='Significance Level')
plt.axvline(test_statistic, color='green', linestyle='--', label='Test Statistic')

# Additional insights
plt.text(test_statistic, 0.02, f'Test Statistic: {test_statistic:.2f}', color='green', ha='center')
plt.text(test_statistic, 0.03, '', color='white', ha='center')  # Create a gap
plt.text(test_statistic, 0.06, f'p-value: {p_value:.6f}', color='black', ha='center')

# Set the labels and title
plt.xlabel('Test Statistic')
plt.ylabel('Probability Density')
plt.title('Hypothesis Test Results')

# Add a legend with larger font size
plt.legend(loc='upper left', fontsize='large')

# Display the plot
plt.show()
```

```python
## 95% CONFIDENCE INTERVAL FOR THE DIFFERENCE IN THE CONVERSION RATE BETWEEN THE TREATMENT AND CONTROL (treatment-control).
## USING THE NORMAL DISTRIBUTION AND UNPOOLED PROPORTIONS FOR THE STANDARD ERROR ##
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample sizes and conversion successes for control and treatment groups
control_size = 24343
control_successes = 955

treatment_size = 24600
treatment_successes = 1139

# Conversion rates for control and treatment groups
control_conversion_rate = control_successes / control_size
treatment_conversion_rate = treatment_successes / treatment_size

# Difference in conversion rates
conversion_rate_diff = treatment_conversion_rate - control_conversion_rate

# Standard error calculation
control_se = math.sqrt(control_conversion_rate * (1 - control_conversion_rate) / control_size)
treatment_se = math.sqrt(treatment_conversion_rate * (1 - treatment_conversion_rate) / treatment_size)
se_difference = math.sqrt(control_se ** 2 + treatment_se ** 2)

# 95% Confidence interval calculation
z_value = 1.96  # For a 95% confidence interval
lower_bound = conversion_rate_diff - z_value * se_difference
upper_bound = conversion_rate_diff + z_value * se_difference

# Plotting the graph
x = np.linspace(conversion_rate_diff - 4 * se_difference, conversion_rate_diff + 4 * se_difference, 1000)
y = norm.pdf(x, conversion_rate_diff, se_difference)

plt.plot(x, y, color='navy', label='Normal Distribution')
plt.fill_between(x, y, where=(x >= lower_bound) & (x <= upper_bound), color='lightblue', alpha=0.4, label='95% Confidence Interval')
plt.plot([lower_bound, lower_bound], [0, max(y)], color='red', linestyle='--', label=f'Lower Bound: {lower_bound:.4f}')
plt.plot([upper_bound, upper_bound], [0, max(y)], color='green', linestyle='--', label=f'Upper Bound: {upper_bound:.4f}')

# Additional insights
plt.text(conversion_rate_diff, max(y) * 0.3, f'Difference in Conversion Rates: {conversion_rate_diff:.4f}', color='black', ha='center')
plt.text(conversion_rate_diff, max(y) * 0.7, '\n\n', color='white', ha='center')  # Add a larger gap
plt.text(conversion_rate_diff, max(y) * 0.2, f'Standard Error: {se_difference:.4f}', color='black', ha='center')

plt.xlabel('Difference in Conversion Rates')
plt.ylabel('Probability Density')
plt.title('Difference in Conversion Rates (Treatment - Control)')

# Add a legend with custom handles and labels
handles = [
    plt.plot([], [], color='navy', label='Normal Distribution')[0],
    plt.fill_between([], [], color='lightblue', alpha=0.4, label='95% Confidence Interval'),
    plt.plot([], [], color='red', linestyle='--', label=f'Lower Bound: {lower_bound:.4f}')[0],
    plt.plot([], [], color='green', linestyle='--', label=f'Upper Bound: {upper_bound:.4f}')[0]
]
labels = ['Normal Distribution', '95% Confidence Interval', f'Lower Bound: {lower_bound:.4f}', f'Upper Bound: {upper_bound:.4f}']
plt.legend(handles=handles, labels=labels, loc='upper right')

plt.show()
```
